# Create differnet groups for analysis

```python
A.Age.describe()
```

```python
group_1 =[]
for i in A["Age"]:
    if i<=25:
        group_1.append("18-25")
    elif i<=30:
        group_1.append("26-30")
    elif i<=35:
        group_1.append("31-35")
    elif i<=40:
        group_1.append("36-40")
    elif i<=45:
        group_1.append("41-45")
    elif i<=50:
        group_1.append("46-50")
    elif i<=60:
        group_1.append("51-60")
    else:
        group_1.append("60+")
```

```python
A["Age_Groups"]=group_1
```

```python
A["DailyRate"].describe()
```

```python
group_2 =[]
for i in A["Age"]:
    if i<=400:
        group_2.append("100-400")
    elif i<=800:
        group_2.append("400-800")
    elif i<=1200:
        group_2.append("800-1200")
    elif i<=1500:
        group_2.append("1200-1500")
    else:
        group_2.append("1500+")
```

```python
A["DailyRate_group"]=group_2
```

```python
A["MonthlyIncome"].describe()
```

```python
group_3 =[]
for i in A["MonthlyIncome"]:
    if i<=5000:
        group_3.append("Less than 5K")
    elif i<=10000:
        group_3.append("5K-10K")
    elif i<=15000:
        group_3.append("10K-15K")
    elif i<=20000:
        group_3.append("15K-20K")
    else:
        group_3.append("20K+")
```

```python
A["MonthlyIncome_group"]=group_3
```

```python
A["MonthlyRate"].describe()
```

```python
group_4 =[]
for i in A["MonthlyIncome"]:
    if i<=5000:
        group_4.append("Less than 5K")
    elif i<=10000:
        group_4.append("5K-10K")
    elif i<=15000:
        group_4.append("10K-15K")
    elif i<=20000:
        group_4.append("15K-20K")
    elif i<=25000:
        group_4.append("20K-25K")
    elif i<=30000:
        group_4.append("25K-30K")
    else:
        group_4.append("30K+")
```

```python
A["MonthlyRate_group"]=group_4
```

```python
A["DistanceFromHome"].describe()
```

```python
group_5 =[]
for i in A["DistanceFromHome"]:
    if i<=5:
        group_5.append("Less than 5KM")
    elif i<=10:
        group_5.append("5KM-10K")
    elif i<=20:
        group_5.append("10KM-20KM")
    elif i<=30:
        group_5.append("20KM-30KM")
    else:
        group_5.append("30KM+")
```

```python
A["DistanceFromHome_group"]=group_5
```
